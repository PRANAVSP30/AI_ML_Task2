This notebook performs systematic data cleaning on the House Prices dataset.
A custom data cleaning module is used to ensure a professional and modular workflow.

```python
import sys
sys.path.append("../src")

from data_cleaner import (
    missing_summary, 
    plot_missing_values, 
    drop_high_missing, 
    impute_numerical, 
    impute_categorical, 
    cleaning_report, 
    save_cleaned_data
)

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/house_prices.csv")
df.head()
```

```python
missing_before = missing_summary(df)
missing_before
```

```python
plot_missing_values(df)
```

```python
df_dropped, dropped_cols = drop_high_missing(df, threshold=60)
df_dropped.head(), dropped_cols
```

```python
df_numeric, numeric_cols = impute_numerical(df_dropped, method="median")
numeric_cols
```

```python
df_cleaned, categorical_cols = impute_categorical(df_numeric)
categorical_cols
```

```python
clean_report = cleaning_report(df, df_cleaned, dropped_cols)
clean_report
```

```python
save_cleaned_data(df_cleaned, "../data/house_prices_cleaned.csv")
```

## Final Cleaning Observations

- Columns with extremely high missing values were removed.
- Numerical values were imputed using median to handle skewed distributions.
- Categorical missing values were imputed using mode.
- Overall missing values reduced from X to Y.
- Cleaned dataset saved as `house_prices_cleaned.csv`.

This structured cleaning workflow ensures the dataset is now ready for Machine Learning.
